2Ô∏è‚É£ Check Perfect Number
‚ùì Question:

A perfect number is equal to the sum of its factors (excluding itself).
Example: 6 ‚Üí 1 + 2 + 3 = 6

```python
num = int(input("Enter number: "))
total = 0

for i in range(1, num):
    if num % i == 0:
        total += i

if total == num:
    print("Perfect number")
else:
    print("Not perfect number")
```

3Ô∏è‚É£ Find Common Elements Without Using in
‚ùì Question:

Find common elements from two lists.

```python
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]

for i in range(len(list1)):
    for j in range(len(list2)):
        if list1[i] == list2[j]:
            print(list1[i])
```

6Ô∏è‚É£ Find GCD (Greatest Common Divisor)
‚ùì Question:

Find GCD of two numbers.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b:
    a, b = b, a % b

print("GCD:", a)
```

7Ô∏è‚É£ Check Strong Number
‚ùì Question:

A strong number = sum of factorial of digits equals number.


```python
num = int(input("Enter number: "))
original = num
total = 0

while num > 0:
    digit = num % 10
    fact = 1
    for i in range(1, digit + 1):
        fact *= i
    total += fact
    num = num // 10

if total == original:
    print("Strong number")
else:
    print("Not strong number")
```

8Ô∏è‚É£ Find Missing Number in List (1 to N)
‚ùì Question:

Find missing number.

```python
numbers = [1, 2, 4, 5]
n = 5

expected_sum = n * (n + 1) // 2
actual_sum = sum(numbers)

print("Missing number:", expected_sum - actual_sum)
```

9Ô∏è‚É£ Convert String to Dictionary
‚ùì Question:

Convert "a=1,b=2,c=3" into dictionary.

```python
text = "a=1,b=2,c=3"
pairs = text.split(",")
result = {}

for pair in pairs:
    key, value = pair.split("=")
    result[key] = int(value)

print(result)
```

üîü Flatten a Nested List
‚ùì Question:

Convert [[1,2],[3,4]] ‚Üí [1,2,3,4]

```python
nested = [[1, 2], [3, 4]]
flat = []

for sublist in nested:
    for item in sublist:
        flat.append(item)

print(flat)
```
