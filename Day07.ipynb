```python
countries = {
    'India': 'New Delhi',
    'USA': 'Washington D.C.',
    'Japan': 'Tokyo',
    'France': 'Paris',
    'Australia': 'Canberra'
}

for country,capital in countries.items():
    print(f"{country}: {capital}")
```

. Set Operations
Create two sets:
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
Perform and print:

Union

Intersection

Difference (A - B)

Symmetric difference

Check if set_a is subset of set_b



```python
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
# 1. Union (all elements from both sets)
union_result = set_a | set_b  # or set_a.union(set_b)
print("1. Union (A âˆª B):", union_result)
# Output: {1, 2, 3, 4, 5, 6, 7, 8}

# 2. Intersection (common elements)
intersection_result = set_a & set_b  # or set_a.intersection(set_b)
print("2. Intersection (A âˆ© B):", intersection_result)
# Output: {4, 5}

# 3. Difference (A - B: elements in A but not in B)
difference_result = set_a - set_b  # or set_a.difference(set_b)
print("3. Difference (A - B):", difference_result)
# Output: {1, 2, 3}

# 4. Symmetric Difference (elements in either set but not both)
symmetric_result = set_a ^ set_b  # or set_a.symmetric_difference(set_b)
print("4. Symmetric Difference (A Î” B):", symmetric_result)
# Output: {1, 2, 3, 6, 7, 8}

# 5. Check if set_a is subset of set_b
is_subset = set_a <= set_b  # or set_a.issubset(set_b)
print("5. Is A subset of B?", is_subset)
# Output: False

# Bonus: Check if set_a is proper subset
is_proper_subset = set_a < set_b
print("6. Is A proper subset of B?", is_proper_subset)
# Output: False

# Bonus: Check if set_b is superset of set_a
is_superset = set_b >= set_a  # or set_b.issuperset(set_a)
print("7. Is B superset of A?", is_superset)
# Output: False
```

. Remove List Duplicates using Sets
Given list: [1, 2, 2, 3, 4, 4, 5, 5, 5, 6, 7, 7]
