âœ… Q3. Create a list of numbers greater than 10 from a list

âœ… Q4. Convert all strings in a list to uppercase

```python
words = ["python", "data", "ai"]
new_word = [word.upper() for word in words]
print(new_word)
```

```python
word= ["python", "data", "ai"]
new_word= [len(word) for word in words]
print(new_word)
```

âœ… Q6. Create a list of vowels from a string

```python
text= "programming"
vowels= [char for char in text if char in "aeiou" ]
print(vowels)
```

âœ… Q7. Remove spaces from a string using comprehension

```python
text = "Hello World Python"
no_spaces = ''.join([char for char in text if char != ' '])
print(no_spaces)
```

âœ… Q8. Create a list of numbers divisible by both 3 and 5

```python
num = [i for i in range(1,51) if i % 3== 0 and i % 5== 0]
print(num)
```

âœ… Q9. Create a dictionary using comprehension (number: square)

```python
squares_dict = {i: i**2 for i in range(1, 6)}
print(squares_dict)
```
