1Ô∏è‚É£ Check if Number is Palindrome (Without Converting to String)
‚ùì Question:

Check palindrome number without using string.

```python
num = int(input("Enter number: "))
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if original == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

2Ô∏è‚É£ Find All Even Numbers in List Using List Comprehension
‚ùì Question:

Extract all even numbers.

```python
numbers = [1, 2, 3, 4, 5, 6]

evens = [num for num in numbers if num % 2 == 0]

print(evens)
```

‚ùì Question:

Find first non-repeating character in string.

```python
text = "aabbcde"

for ch in text:
    if text.count(ch) == 1:
        print("First non-repeating:", ch)
        break
```

4Ô∏è‚É£ Move All Zeros to End of List
‚ùì Question:

Move all zeros to end without changing order.

```python
numbers = [0, 1, 0, 3, 12]

result = [num for num in numbers if num != 0]
zeros = [0] * numbers.count(0)

print(result + zeros)
```

5Ô∏è‚É£ Find Maximum Difference in List
‚ùì Question:

Find max difference between two elements (later element must be bigger).

```python
numbers = [2, 3, 10, 6, 4, 8, 1]

min_value = numbers[0]
max_diff = 0

for num in numbers:
    if num < min_value:
        min_value = num
    elif num - min_value > max_diff:
        max_diff = num - min_value

print("Maximum difference:", max_diff)
```

6Ô∏è‚É£ Check if Two Lists are Equal (Without ==)
‚ùì Question:

Compare two lists manually.

```python
list1 = [1, 2, 3]
list2 = [1, 2, 3]

equal = True

if len(list1) != len(list2):
    equal = False
else:
    for i in range(len(list1)):
        if list1[i] != list2[i]:
            equal = False
            break

print("Equal lists:", equal)
```

7Ô∏è‚É£ Find Sum of Even and Odd Separately
‚ùì Question:

Find sum of even and odd numbers separately.

```python
numbers = [1, 2, 3, 4, 5, 6]

even_sum = 0
odd_sum = 0

for num in numbers:
    if num % 2 == 0:
        even_sum += num
    else:
        odd_sum += num

print("Even sum:", even_sum)
print("Odd sum:", odd_sum)
```

üîü Rotate List by K Positions
‚ùì Question:

Rotate list right by k positions.

```python
numbers = [1, 2, 3, 4, 5]
k = 2

k = k % len(numbers)

rotated = numbers[-k:] + numbers[:-k]

print(rotated)
```
