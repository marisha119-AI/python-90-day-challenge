â“ Question:

Find two numbers whose sum equals target (optimized using dictionary).

```python
numbers = [2, 7, 11, 15]
target = 9

seen = {}

for i, num in enumerate(numbers):
    diff = target - num
    if diff in seen:
        print("Pair:", diff, num)
        break
    seen[num] = i
```

2ï¸âƒ£ Find Longest Substring Without Repeating Characters
â“ Question:

Find length of longest substring without repeating characters.

```python
text = "abcabcbb"
seen = set()
left = 0
max_length = 0

for right in range(len(text)):
    while text[right] in seen:
        seen.remove(text[left])
        left += 1
    seen.add(text[right])
    max_length = max(max_length, right - left + 1)

print("Longest length:", max_length)
```
