1Ô∏è‚É£ Check Armstrong Number
‚ùì Question:

Check whether a number is an Armstrong number.

```python
num = int(input("Enter number: "))
original = num
result = 0

while num > 0:
    digit = num % 10
    result += digit ** 3
    num = num // 10

if original == result:
    print("Armstrong number")
else:
    print("Not Armstrong number")
```

2Ô∏è‚É£ Find Minimum in List (Without min())
‚ùì Question:

Find smallest number in list without using min().

```python
numbers = [45, 12, 78, 3, 25]

smallest = numbers[0]

for num in numbers:
    if num < smallest:
        smallest = num

print("Smallest:", smallest)
```

3Ô∏è‚É£ Count Uppercase and Lowercase Letters
‚ùì Question:

Count uppercase and lowercase letters in string.

```python
text = input("Enter text: ")
upper = 0
lower = 0

for ch in text:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1

print("Uppercase:", upper)
print("Lowercase:", lower)
```

Print first 10 Fibonacci numbers.

```python
a = 0
b = 1

for i in range(10):
    print(a)
    a, b = b, a + b
```

5Ô∏è‚É£ Remove Duplicates from List
‚ùì Question:

Remove duplicate values.

```python
numbers = [1, 2, 2, 3, 4, 4, 5]
unique = []

for num in numbers:
    if num not in unique:
        unique.append(num)

print(unique)
```

7Ô∏è‚É£ Check Leap Year
‚ùì Question:

Check if year is leap year.

```python
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not Leap Year")
```

8Ô∏è‚É£ Sort List Without sort()
‚ùì Question:

Sort list manually.

```python
numbers = [5, 2, 8, 1, 3]

for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] > numbers[j]:
            numbers[i], numbers[j] = numbers[j], numbers[i]

print(numbers)
```

9Ô∏è‚É£ Count Frequency of Element
‚ùì Question:

Count how many times 5 appears.

```python
numbers = [5, 1, 5, 3, 5, 2]

count = 0
for num in numbers:
    if num == 5:
        count += 1

print("Frequency of 5:", count)
```

üîü Find Sum of Digits
‚ùì Question:

Find sum of digits in number.

```python
num = int(input("Enter number: "))
total = 0

while num > 0:
    digit = num % 10
    total += digit
    num = num // 10

print("Sum of digits:", total)
```
