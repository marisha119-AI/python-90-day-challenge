 Q1. Write a function to return the largest number in a list.

```python
def find_largest(numbers):
    return max(numbers)

print(find_largest([10, 25, 7, 40, 15]))
```

Q2. Write a function to count vowels in a string.

```python
def count_vowels(text):
    vowels = "aeiou"
    count = 0

    for char in text.lower():
        if char in vowels:
            count += 1

    return count

print(count_vowels("Python Programming"))
```

 Q3. Write a program to remove duplicates from a list.

```python
numbers = [1, 2, 2, 3, 4, 4, 5]

unique_numbers = list(set(numbers))

print(unique_numbers)
```

Q4. Write a function that returns the square of numbers in a list.

```python
def square_list(numbers):
    return [num**2 for num in numbers]

print(square_list([1,2,3,4]))
```

Q5. Write a program to create a dictionary of number and its square.

```python
numbers = [1,2,3,4]

squares = {num: num**2 for num in numbers}

print(squares)
```

Write a function to reverse a string.

```python
def reverse_string(text):
    return text[::-1]

print(reverse_string("Python"))
```

 Q7. Write a function that checks if a number is even.

```python
def is_even(num):
    return num % 2 == 0

print(is_even(10))
```

Q8. Write a function that sums all numbers in a list.

```python
def sum_list(numbers):
    return sum(numbers)

print(sum_list([1,2,3,4,5]))
```

Q9. Write a program to merge two lists.

```python
list1 = [1,2,3]
list2 = [4,5,6]

merged = list1 + list2

print(merged)
```

Q10. Write a function that prints key-value pairs from a dictionary.

```python
def print_dict(data):
    for key, value in data.items():
        print(key, ":", value)

student = {"name":"Marisha", "age":21}

print_dict(student)
```
