Write a Python program to count the number of vowels in a given string.

```python
text = input("Enter a string")
vowels = "aeiouAEIOU"
count = 0
for char in text:
    if char in vowels:
        count += 1
print(count)       
```

Q7. Write a Python program to replace all vowels in a string with *.

```python
text = input("Enter a string: ")
vowels = "aeiouAEIOU"

result = ""
for char in text:
    if char in vowels:
        result += "*"
    else:
        result += char

print("Result:", result)
        
```
