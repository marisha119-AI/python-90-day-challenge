Recursion is a technique where a function calls itself to solve a problem.

‚úÖ Q2. Write a recursive function to print numbers from 1 to N

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)

print_numbers(5)
```

‚úÖ Q3. Write a recursive function to find the factorial of a number

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

‚úÖ Q4. Write a recursive function to find the sum of numbers from 1 to N

```python
def sum_to_n(n):
    if n == 0:
        return 0
    return n + sum_to_n(n - 1)

print(sum_to_n(5))
```

‚úÖ Q5. Write a recursive function to print a string in reverse

```python
def reverse_string(text):
    if text == "":
        return text
    return reverse_string(text[1:]) + text[0]

print(reverse_string("python"))
```

‚úÖ Q6. Write a recursive function to calculate power (a‚Åø)

```python
def power(a, n):
    if n == 0:
        return 1
    return a * power(a, n - 1)

print(power(2, 3))
```

‚úÖ Q7. Write a recursive function to count digits in a number

```python
def count_digits(num):
    if num == 0:
        return 0
    return 1 + count_digits(num // 10)

print(count_digits(12345))
```

‚úÖ Q8. Write a recursive function to check if a number is even or odd

```python
def is_even(num):
    if num == 0:
        return "Even"
    return is_even(num - 2)

print(is_even(6))
```

‚úÖ Q9. What is a base case in recursion?

A base case is the condition that stops recursion, preventing infinite calls.

‚úÖ Q10. Write a recursive function to find the maximum number in a list

```python
def find_max(lst):
    if len(lst) == 1:
        return lst[0]
    return max(lst[0], find_max(lst[1:]))

print(find_max([3, 7, 2, 9, 5]))
```
