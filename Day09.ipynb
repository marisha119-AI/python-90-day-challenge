âœ… Q2. Find the sum of digits of a number

```python
num = int(input("Enter a number: "))
total = 0

while num > 0:
    total += num % 10
    num //= 10

print("Sum of digits:", total)
```

âœ… Q3. Reverse a number using a loop

```python
num = int(input("Enter a number: "))
reverse = 0

while num > 0:
    reverse = reverse * 10 + num % 10
    num //= 10

print("Reversed number:", reverse)
```

âœ… Q4. Check whether a number is an Armstrong number

```python
num = int(input("Enter a number: "))
temp = num
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** 3
    temp //= 10

if total == num:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

âœ… Q5. Print the first 10 prime numbers

```python
count = 0
num = 2

while count < 10:
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num)
        count += 1
    num += 1
```

âœ… Q6. Multiplication table using a while loop

```python
num = int(input("Enter a number: "))
i = 1

while i <= 10:
    print(num, "x", i, "=", num * i)
    i += 1
```

âœ… Q7. Find the largest number in a list

```python
numbers = [10, 25, 3, 45, 7]
largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest number:", largest)
```

âœ… Q8. Remove duplicate elements from a list

```python
numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = []

for num in numbers:
    if num not in unique_numbers:
        unique_numbers.append(num)

print("List without duplicates:", unique_numbers)
```

âœ… Q9. Count how many times a digit appears in a number

```python
num = input("Enter a number: ")
digit = input("Enter digit to count: ")

count = 0
for ch in num:
    if ch == digit:
        count += 1

print("Digit appears", count, "times")
```

âœ… Q10. Keep taking numbers until -1, then print sum

```python
total = 0

while True:
    num = int(input("Enter a number (-1 to stop): "))
    if num == -1:
        break
    total += num

print("Sum of entered numbers:", total)
```
