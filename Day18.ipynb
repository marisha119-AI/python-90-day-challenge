Write a function generate_prompt() that takes a user role and task as parameters and returns a formatted prompt string.

Q2.

Write a function that accepts any number of instructions using *args and prints them one by one.

```python
def instructions(*args):
    for items in args:
        print(items)
instructions("Think clearly", "Write code", "Test often")        
```

Q3.

Write a function that accepts keyword arguments (**kwargs) like tone, language, and length, and prints their values.

```python
def argu(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")
        
argu(tone="Friendly", language="English", length="Short")     
```

Q5.

Create a logging decorator that prints:
Function <name> is called

```python
def log_name(func):
    def wrapper():
        print(f"Function {func.__name__} is called")
        
        func()
        
    return wrapper

@log_name
def say_hi():
    print("Hi!")

say_hi()
```

Q6.

Use the logging decorator from Q5 on a function called create_prompt().

```python
# 1. Define the decorator
def log_name(func):
    def wrapper(*args, **kwargs):
        print(f"Function {func.__name__} is called")
        return func(*args, **kwargs)
    return wrapper

@log_name
def create_prompt(task, persona):
    return f"As a {persona}, please {task}."

result = create_prompt("write a poem", "Pirate")
print(result)
```

Q7.

Write a validation decorator that checks if input text is empty.

If empty â†’ print "Invalid prompt"

Else â†’ run the function

```python
def check_empty(func):
    def wrapper(text):
        if text == "":
            print("Invalid prompt")
        else:
            func(text)
    return wrapper

@check_empty
def start_ai(text):
    print("AI is starting with:", text)

start_ai("")       # Output: Invalid prompt
start_ai("Hello")  # Output: AI is starting with: Hello
```

Q8.

Apply the validation decorator to a function that prints a user prompt.

```python
def check_empty(func):
    def wrapper(text):
        if text == "":
            print("Invalid prompt")
        else:
            func(text)
    return wrapper

@check_empty
def print_user_prompt(text):
    print(f"User said: {text}")

print_user_prompt("")       # This will trigger "Invalid prompt"
print_user_prompt("Hello!") # This will work and print "User said: Hello!"
```
