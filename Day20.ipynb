1Ô∏è‚É£ Find Duplicate Elements in a List
‚ùì Question:

Find duplicate elements from a list.

```python
numbers = [1, 2, 3, 2, 4, 5, 1]
duplicates = []

for num in numbers:
    if numbers.count(num) > 1 and num not in duplicates:
        duplicates.append(num)

print("Duplicates:", duplicates)
```

3Ô∏è‚É£ Find All Prime Numbers in Range
‚ùì Question:

Print prime numbers between 1 and 50.

```python
for num in range(2, 51):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num)
```

4Ô∏è‚É£ Find Longest Word in Sentence
‚ùì Question:

Find longest word in a sentence.

```python
text = "Python is very powerful language"
words = text.split()

longest = words[0]

for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)
```

6Ô∏è‚É£ Find Second Largest Without Built-in Functions
‚ùì Question:

Find second largest number manually.

```python
numbers = [10, 20, 5, 40, 30]

largest = second = float('-inf')

for num in numbers:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num

print("Second largest:", second)
```

7Ô∏è‚É£ Count Words Frequency
‚ùì Question:

Count frequency of each word.

```python
text = "python is easy and python is powerful"
words = text.split()
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

print(freq)
```

8Ô∏è‚É£ Check if List is Sorted
‚ùì Question:

Check whether list is sorted in ascending order.

```python
numbers = [1, 2, 3, 4, 5]

is_sorted = True

for i in range(len(numbers) - 1):
    if numbers[i] > numbers[i + 1]:
        is_sorted = False
        break

if is_sorted:
    print("List is sorted")
else:
    print("Not sorted")
```

9Ô∏è‚É£ Find Intersection of Two Lists
‚ùì Question:

Find common elements using set.

```python
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]

common = list(set(list1) & set(list2))

print("Intersection:", common)
```

üîü Generate Random Password
‚ùì Question:

Generate 8-character random password.

```python
import random
import string

characters = string.ascii_letters + string.digits
password = ""

for i in range(8):
    password += random.choice(characters)

print("Password:", password)
```
