Q 60. 11. Matrix Operations
python
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
Add matrices A and B

Transpose matrix A (swap rows and columns)

Multiply matrix by scalar (multiply all elements by 2)



```python
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
addition = [
    [A[i][j] + B[i][j] for j in range(len(A[0]))]
    for i in range(len(A))
]

print("Matrix Addition:", addition)
```

```python
transpose_A = [
    [A[j][i] for j in range(len(A))]
    for i in range(len(A[0]))
]

print("Transpose of A:", transpose_A)
```

```python
scalar_multiply = [
    [element * 2 for element in row]
    for row in A
]

print("Matrix A multiplied by 2:", scalar_multiply)
```
