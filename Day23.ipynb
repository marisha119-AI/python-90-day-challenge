3ï¸âƒ£ Find Leaders in List
â“ Question:

A leader is greater than all elements to its right.

```python
numbers = [16, 17, 4, 3, 5, 2]

max_from_right = numbers[-1]
print(max_from_right)

for i in range(len(numbers) - 2, -1, -1):
    if numbers[i] > max_from_right:
        max_from_right = numbers[i]
        print(max_from_right)
```

4ï¸âƒ£ Find First Duplicate Element
â“ Question:

Find first repeating element in list.

```python
numbers = [2, 1, 3, 5, 3, 2]

seen = set()

for num in numbers:
    if num in seen:
        print("First duplicate:", num)
        break
    seen.add(num)
```

6ï¸âƒ£ Find Longest Common Prefix
â“ Question:

Find common starting letters in list of words.

Example: ["flower", "flow", "flight"] â†’ "fl"

```python
words = ["flower", "flow", "flight"]

prefix = words[0]

for word in words[1:]:
    while not word.startswith(prefix):
        prefix = prefix[:-1]

print("Longest common prefix:", prefix)
```

7ï¸âƒ£ Count Pairs with Given Difference
â“ Question:

Count pairs where difference = 2.

```python
numbers = [1, 5, 3, 4, 2]
difference = 2
count = 0

for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if abs(numbers[i] - numbers[j]) == difference:
            count += 1

print("Total pairs:", count)
```

8ï¸âƒ£ Check Balanced Parentheses
â“ Question:

Check if parentheses are balanced.

Example: "((()))" â†’ Balanced

```python
text = "((()))"
stack = []

for ch in text:
    if ch == "(":
        stack.append(ch)
    else:
        if stack:
            stack.pop()
        else:
            print("Not Balanced")
            break

if not stack:
    print("Balanced")
```

â“ Question:

Find element appearing more than n/2 times.

```python
numbers = [3, 3, 4, 2, 3, 3, 3]
n = len(numbers)

for num in numbers:
    if numbers.count(num) > n // 2:
        print("Majority element:", num)
        break
```
