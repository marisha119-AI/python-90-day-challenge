1Ô∏è‚É£ Find Pair with Given Sum
‚ùì Question:

Find two numbers in list whose sum equals target.

```python
numbers = [2, 7, 11, 15]
target = 9

for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] + numbers[j] == target:
            print("Pair:", numbers[i], numbers[j])
```

3Ô∏è‚É£ Find Largest Word Length
‚ùì Question:

Find length of longest word in sentence.

```python
text = "Python makes problem solving easier"
words = text.split()

max_length = 0

for word in words:
    if len(word) > max_length:
        max_length = len(word)

print("Longest word length:", max_length)
```

4Ô∏è‚É£ Remove Duplicate Characters from String
‚ùì Question:

Remove duplicate characters.

```python
text = "programming"
result = ""

for ch in text:
    if ch not in result:
        result += ch

print(result)
```

5Ô∏è‚É£ Find Intersection Without Set
‚ùì Question:

Find common elements without using set.

```python
list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]

common = []

for num in list1:
    if num in list2 and num not in common:
        common.append(num)

print(common)
```

6Ô∏è‚É£ Count Even & Odd Digits in Number
‚ùì Question:

Count even and odd digits separately.

```python
num = int(input("Enter number: "))
even = 0
odd = 0

while num > 0:
    digit = num % 10
    if digit % 2 == 0:
        even += 1
    else:
        odd += 1
    num = num // 10

print("Even digits:", even)
print("Odd digits:", odd)
```

7Ô∏è‚É£ Check if List is Palindrome
‚ùì Question:

Check whether a list is palindrome.

```python
numbers = [1, 2, 3, 2, 1]

if numbers == numbers[::-1]:
    print("Palindrome list")
else:
    print("Not palindrome")
```

8Ô∏è‚É£ Find Element with Maximum Frequency
‚ùì Question:

Find most frequent element.

```python
numbers = [1, 2, 2, 3, 3, 3, 4]

max_count = 0
most_frequent = None

for num in numbers:
    count = numbers.count(num)
    if count > max_count:
        max_count = count
        most_frequent = num

print("Most frequent:", most_frequent)
```

9Ô∏è‚É£ Create Dictionary from Two Lists
‚ùì Question:

Create dictionary using two lists.

```python
keys = ["a", "b", "c"]
values = [1, 2, 3]

result = {}

for i in range(len(keys)):
    result[keys[i]] = values[i]

print(result)
```

üîü Count Consonants in String
‚ùì Question:

Count consonants in string.

```python
text = input("Enter text: ")
count = 0

for ch in text.lower():
    if ch.isalpha() and ch not in "aeiou":
        count += 1

print("Consonants:", count)
```
