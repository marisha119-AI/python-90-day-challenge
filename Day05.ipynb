Q 46. Write a function that returns the sum of numbers from 1 to N.

```python
def sum(n):
    total = 0
    for i in range(1, n+ 1):
        total += i
    return total
sum(10)      
```

Q 47. Write a function that checks whether a given year is a leap year.

```python
def leap_year(n):
    if n % 4== 0 and n % 400 == 0:
        print(n, "Year is Leap year")
    elif n% 100 != 0:
        print(n, "Year is Not Leap year")

leap_year(2000)
```

Q 49. Create a function count_vowels(text) that counts how many vowels (a, e, i, o, u) are in a given string.



```python
def vowels(text):
    vowels= "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count
vowels("Marisha Dwivedi")    
```

Q 50. Write a function that takes a string and returns the string in reverse order.

```python
def reverse(text):
    text= text[::-1]
    return text
reverse("Marisha Dwivedi")    
```
