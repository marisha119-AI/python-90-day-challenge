Q 36. Write a Python program to count the number of digits in a given number using a loop.

```python
num= int(input("Enter a number:"))
count = 0
while num > 0:
    num= num // 10
    count += 1
print(count)
```

Q 37. Write a Python program to check whether a given number is a prime number.


```python
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a Prime number")
else:
    is_prime = True
    
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime number")
    else:
        print("Not a Prime number")
```

Q 38. Write a Python program to print the Fibonacci series up to N terms.

```python
n = int(input("Enter number of terms: "))

a = 0
b = 1

print("Fibonacci Series:")

for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```

Q 39. Write a Python program that keeps asking the user to enter a number until they type 0, then print the sum of all entered numbers.

```python
total = 0

while True:
    num = int(input("Enter a number (0 to stop): "))
    
    if num == 0:
        break
    
    total += num

print("Sum of entered numbers:", total)
```

Q 40. Write a Python program to calculate the factorial of a number using a loop.

```python
num = int(input("Enter a number: "))

factorial = 1

for i in range(1, num + 1):
    factorial *= i

print("Factorial =", factorial)
```
