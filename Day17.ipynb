âœ… Q1. What is a lambda function in Python?

A lambda function is a small anonymous function written in one line.

```python
square = lambda x: x * x
print(square(5))
```

âœ… Q2. Write a lambda function to add two numbers

```python
add = lambda a, b: a + b
print(add(10, 20))
```

âœ… Q3. What is the map() function?

map() applies a function to each element of a list.

âœ… Q4. Use map() to find squares of a list

```python
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x * x, numbers))
print(squares)
```

âœ… Q5. What is the filter() function?

filter() selects elements from a list based on a condition.

âœ… Q6. Use filter() to get even numbers from a list

```python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

âœ… Q7. What is the reduce() function?

reduce() reduces a list to a single value by applying a function repeatedly.

âœ… Q8. Use reduce() to find the sum of a list

```python
from functools import reduce

numbers = [1, 2, 3, 4]
total = reduce(lambda a, b: a + b, numbers)
print(total)
```

âœ… Q9. Find the maximum number in a list using reduce()

```python
from functools import reduce

numbers = [10, 25, 5, 40]
max_num = reduce(lambda a, b: a if a > b else b, numbers)
print(max_num)
```

âœ… Q10. Combine map() and filter()

```python
numbers = [1, 2, 3, 4, 5, 6]

result = list(
    map(lambda x: x * x,
        filter(lambda x: x % 2 == 0, numbers))
)

print(result)
```
